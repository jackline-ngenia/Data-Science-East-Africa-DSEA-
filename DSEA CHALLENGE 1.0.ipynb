## 4. Draw any two plots of your choice.

```python
# importing matplotlib
import matplotlib.pyplot as plt
```

```python
# multiple groupby both Satisfcation and Gender on the basis of Importance
df = mobile_banking.groupby(['Satisfaction', 'Gender']).count()['Importance']

# plot the result
df.unstack().plot()
plt.show()
```

```python
# multiple groupby both MbSafety and Gender on the basis of use
df = mobile_banking.groupby(['MbSafety', 'Gender']).count()['use']

# plot the result
df.unstack().plot()
plt.show()
```
